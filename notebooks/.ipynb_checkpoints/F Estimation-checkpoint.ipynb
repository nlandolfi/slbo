```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg
```

```python
F = np.tril(np.random.rand(10, 10))
plt.imshow(F)
plt.title("F")
plt.figure()
print(F)
plt.title('FF^T')
plt.imshow(np.matmul(F, F.T))
```

```python
ws = []
ys = []
for i in range(int(1e3)):
    w = np.random.randn(10)
    ws.append(w)
    ys.append(np.matmul(F, w))
```

```python
est = np.zeros((10,10))
Y = np.hstack(ys)
ymean = np.mean(Y)
for y in ys:
    est += np.outer(y-ymean, y-ymean)
print(est)
est /= float(len(ys))
```

```python
Fhat = np.linalg.cholesky(est)
```

```python
print(f"{np.linalg.norm(Fhat - F)}")
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0, 0].imshow(F)
ax[0, 0].set_title("F")
ax[0, 1].imshow(Fhat)
ax[0, 1].set_title("Fhat")
ax[1, 0].imshow(np.matmul(F, F.T))
ax[1, 0].set_title("FF^T")
ax[1, 1].imshow(est)
ax[1, 1].set_title("FF^That")
print(f"{np.linalg.norm(Fhat - F)}")
print(f"{np.linalg.norm(est - np.matmul(F, F.T))}")
```

```python
for i in range(5):
    w = np.random.randn(10)
    y = np.matmul(F, w)
    plt.stem(y)
    plt.figure()
    west = np.linalg.solve(Fhat, y)
    plt.scatter(np.arange(10), west)
    plt.scatter(np.arange(10), w, color='red')
    plt.figure()
```
