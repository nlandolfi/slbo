```python
real_warm = []
virt_warm = []
virt_std = []
warm = False
already = False
for line in lines:
    if "pre-warm-up" == line["args"][0]:
        warm = True
        if already:
            break
    if "post-warm-up" == line["args"][0]:
        warm = False
        already = True
    if "iteration" == line["args"][0]:
        if warm:
            if line["args"][1] == "Real Env":
                real_warm.append(line["args"][3]) 
            if line["args"][1] == "Virt Env":
                virt_warm.append(line["args"][3])
                virt_std.append(line["args"][4])
```

```python
warmup_rewards = []
pre_warmup_real_reward = 0.0
post_warmup_real_reward = 0.0
warmup_real_rewards = []
warmup_virt_rewards = []
in_warmup = False

slbo_rewards = []
trpo_rewards = []

for line in lines:
    if line["args"][0] == "pre-warm-up":
        in_warmup = True
        if line["args"][1] == "Real Env":
            pre_warmup_real_reward = line["args"][3]
    if line["args"][0] == "post-warm-up":
        in_warmup = False
        if line["args"][1] == "Real Env":
            post_warmup_real_reward = line["args"][3]
    if line["args"][0] == "iteration":
        if line["args"][1] == "Real Env":
            if in_warmup:
                warmup_real_rewards.append(line["args"][3])
        if line["args"][1] == "Virt Env":
            if in_warmup:
                warmup_virt_rewards.append(line["args"][3])
    if "[TRPO]" in line["fmt"]:
        if in_warmup:
            warmup_rewards.append(line["args"][2])
        else:
            slbo_rewards.append(line["args"][2])
        trpo_rewards.append(line["args"][2])
        
print(warmup_real_rewards)
```

```python
WARMUP_N_ITERS = 40
WARMUP_N_MODEL_ITERS = 100
WARMUP_N_POLICY_ITERS = 40
WARMUP_N_EVALUATE_ITERS = 1
```
